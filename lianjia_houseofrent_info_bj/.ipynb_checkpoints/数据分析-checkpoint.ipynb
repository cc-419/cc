```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
import statsmodels.api as sm
import scipy 
data_c1=pd.read_csv('北京两室租房房源信息_1.csv')
data_c2=pd.read_csv('北京两室租房房源信息_2.csv')
```

```python
data_c1=pd.read_csv('北京两室租房房源信息_1.csv')
data_c2=pd.read_csv('北京两室租房房源信息_2.csv')

## 数据合并
data_c3=pd.merge(data_c1,data_c2,how='inner',on='house_code',suffixes=('_left', '_right'))

##数据处理，
data_c4=data_c3[(data_c3['beadroom_num']==2)&(data_c4['storey_num']!='')] ## 过滤脏数据
data_c4['area_1']=data_c4['area'].str.extract(r'(\d*)').astype(int) ## 因为area字段有非数值的，仅提取数值
data_c4.drop(columns=['area'],inplace=True) ## 删除area列
data_c4['orientations']=data_c4['orientations'].str.replace(' ','') ## 替换方向字段里的控制
storey_tmp=data_c4['storey'].str.split(r'\/|层',expand=True) ## storey字段按特定字段分列，
storey_tmp.columns=['storey_type','-','storey_num','-'] ## 重置列名
data_c4=data_c4.join(storey_tmp[['storey_type','storey_num']]) ## 合并数据
data_c4.drop(columns=['area','storey'],inplace=True,errors='ignore')
data_c4['area_range']=pd.cut(data_c4['area_1'],bins=[32,50,80,100,120,316],right=False,precision=0,include_lowest=True) ## 按桶分割数据
data_c4['longitude']=data_c4['longitude'].str.replace(',','').astype(float) ##转化数据类型
data_c4['storey_num']=data_c4['storey_num'].astype(int) ##转化数据类型
data_c4.info()


### 脏数据 BJ2572351939857883136
```

```python
data_c4.groupby(['house_condition_name']).agg({'house_code':np.size,
                                               'rent_price':np.mean,'area_1':np.mean}).sort_values(['house_code','rent_price'],ascending=[False,False])
```

```python
##价格VS面积
fig_c,ax_c=plt.subplots(1,2,figsize=(16,5))

x_s_data=data_c4[data_c4['area_1']<100]['area_1']
y_s_data=data_c4[data_c4['area_1']<100]['rent_price']

r_area_prince,p_area_prince=scipy.stats.pearsonr(x_s_data, y_s_data)
print('租金VS面积(面积100平以内)的相关性是{},相关性检验P-value是{}'.format(r_area_prince,p_area_prince))

sns.set()
ax_c[0].scatter(x_s_data,y_s_data)
ax_c[0].set_title('rent_price Vs area(area<100)')
ax_c[0].set_xlabel('area')
ax_c[0].set_ylabel('rent_price')
ax_c[0].tick_params(labelsize = 10)


### 各面积区间箱线图
data_1=data_c4[(data_c4['area_1']>=32)&(data_c4['area_1']<50)]['rent_price']
data_2=data_c4[(data_c4['area_1']>=50)&(data_c4['area_1']<80)]['rent_price']
data_3=data_c4[(data_c4['area_1']>=80)&(data_c4['area_1']<100)]['rent_price']
data_4=data_c4[(data_c4['area_1']>=100)&(data_c4['area_1']<120)]['rent_price']
data_5=data_c4[(data_c4['area_1']>=120)&(data_c4['area_1']>=120)]['rent_price']
label=['[32, 50)','[50, 80)','[80, 100)','[100, 120)','[120, 316)']
ax_c[1].boxplot(x=[data_1,data_2,data_3,data_4,data_5],labels=label)
#ax_c[1]=sns.boxplot(x =data_c4[['area_range']] , y = data_c4[['rent_price']], data = data_c4) 
plt.show()
print('结论:\n（1）租金和面积并没有很强的正相关性;\n（2）;\n')
```

```python
fig_c2,ax_c2=plt.subplots(2,4,figsize=(20,10))

ax_c2[0][0].hist('rent_price',bins=25,density=True,data=data_c4[data_c4['orientations']=='东西'])
ax_c2[0][0].set_title('the median of east direction:{}'.format(data_c4[data_c4['orientations'].isin(['东西'])]['rent_price'].median()))

ax_c2[0][1].hist('rent_price',bins=25,density=True,data=data_c4[data_c4['orientations'].isin(['南北','南'])])
ax_c2[0][1].set_title('the median of south direction:{}'.format(data_c4[data_c4['orientations'].isin(['南北','南'])]['rent_price'].median()))

ax_c2[0][2].hist('rent_price',bins=25,density=True,data=data_c4[data_c4['civil_water'].isin(['民水'])])
ax_c2[0][2].set_title('the median of is_civil_water:{}'.format(data_c4[data_c4['civil_water'].isin(['民水','民水'])]['rent_price'].median()))

ax_c2[0][3].hist('rent_price',bins=25,density=True,data=data_c4[data_c4['civil_water'].isin(['商水'])])
ax_c2[0][3].set_title('the median of is_not_civil_water:{}'.format(data_c4[data_c4['civil_water'].isin(['商水','商水'])]['rent_price'].median()))


ax_c2[1][0].hist('rent_price',bins=25,density=True,data=data_c4[data_c4['lease_term'].isin(['3年以上'])])
ax_c2[1][0].set_title('the median of lease_term_up3year:{}'.format(data_c4[data_c4['lease_term'].isin(['3年以上','3年以上'])]['rent_price'].median()))

ax_c2[1][1].hist('rent_price',bins=25,density=True,data=data_c4[data_c4['lease_term'].isin(['1年以内','1~3个月'])])
ax_c2[1][1].set_title('the median of lease_term_in1year:{}'.format(data_c4[data_c4['lease_term'].isin(['1年以内','1~3个月'])]['rent_price'].median()))

ax_c2[1][2].hist('rent_price',bins=25,density=True,data=data_c4[data_c4['lease_term'].isin(['3年以内','2年以内'])])
ax_c2[1][2].set_title('the median of lease_term_in3year:{}'.format(data_c4[data_c4['lease_term'].isin(['2年以内','3年以内'])]['rent_price'].median()))





plt.show()

```

```python

fig_c3,ax_c3=plt.subplots(1,2,figsize=(20,10))

plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

data_storeynum_cc=pd.pivot_table(data_c4[data_c4['storey_num']!='']
               ,index=['storey_num'],values=['rent_price'],aggfunc=[np.mean,np.median,np.size]).reset_index()
data_storeynum_cc.columns=['storey_num','mean_rent_price','median_rent_price','house_num']
data_storeynum_cc['storey_num']=data_storeynum_cc['storey_num'].astype(int)
data_storeynum_cc.sort_values(by='storey_num',ascending=True,inplace=True)


ax_c3[0].plot(data_storeynum_cc['storey_num'],data_storeynum_cc['mean_rent_price'],marker='o',lw=3)
ax_c3[0].plot(data_storeynum_cc['storey_num'],data_storeynum_cc['median_rent_price'],marker='*',lw=3)
ax_c3[0].legend(['mean_rent_price','median_rent_price'],loc=0)

ax_c3_tmp=ax_c3[0].twinx()
ax_c3_tmp.bar(data_storeynum_cc['storey_num'],data_storeynum_cc['house_num'])

ax_c3[1].bar(data_storeynum_cc['storey_num'],data_storeynum_cc['house_num'])



plt.show()
```
