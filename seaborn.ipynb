## step1 处理数据

```python
##鸢尾花数据处理
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn import datasets ##sklearn自带数据，但需要做处理
Iris = datasets.load_iris()
labels=['sepal_length','sepal_width','petal_length','petal_width']
iris=pd.DataFrame(Iris.data, columns=labels)
target_data=pd.DataFrame(Iris.target, columns=['target'])
iris=iris.join(target_data)
iris["species"] =iris["target"]
iris.loc[iris["species"]==0,'species']='setosa'
iris.loc[iris["species"]==1,'species']='versicolor'
iris.loc[iris["species"]==2,'species']='virginica'
# target_map={0:'setosa',1:'versicolor',2:'virginica'}
# iris['target_name2']=iris['target'].map(target_map)
iris.head()
```

```python
# 散点图
sns.relplot(x="sepal_length", y="sepal_width", data=iris)
```

```python
## 不同类别花的散点图
sns.relplot(x="sepal_length", y="sepal_width", hue="species", data=iris)
```

```python
## style 参数可以赋予不同类别的散点不同的形状
sns.relplot(x="sepal_length", y="sepal_width",
            hue="species", style="species", data=iris)
```

```python
## 不只是散点图，该方法还支持线形图，只需要指定 kind="line" 参数即可，
sns.relplot(x="sepal_length", y="petal_length",
            hue="species", style="species", kind="line", data=iris)
```

```python
## 线性图直接用lineplot也可以
sns.lineplot(x="sepal_length", y="petal_length",
             hue="species", style="species", data=iris)
```
