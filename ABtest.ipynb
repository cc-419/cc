### 为什么做AB？
- A1：是实现因果推断的一种有效方法，在所有条件都相等的条件下，只改变其中一个条件，对比两者产生的的效果的不同，AB提倡的是科学有效性
- A2: ABtest是为了同一个目标制定两个方案，在同一个时间维度下，分别让相似群体随机使用同一个方案，收集各群组的用户数据，后根据显著性检验分析评估好的方案
### 什么是AA实验，为什么做AA？
按照某种随机的方法把人群，时间或者空间随机分成两组，每组都上同样的策略。AA可以检测实验随机化是否均匀，往往是AB实验 准确性的一种可靠的保证
### 什么是假设检验
首先对总体参数提出某种假设，然后利用样本信息去判断这个假设是否成立对过程，一般提出两种假设，原假设H0 和备则假设H1，原假设是我们要收集数据去反对的假设
### 什么是第一类错误和第二类错误
第一类错误:当H0为真，我们拒绝H0，这类错误为第一类错误，发生的概率是𝜶，也就是误报效果，一般设置为0.05
第二类错误:当H0为假，我们接受H0，这类错误为第二类错误，发生的概率是𝜷，也就是未能识别效果，一般设置为0.2
### 第一类错误和第二类错误的关系
𝜶 和 𝜷 的关系是此消彼长。一般来说，当其他条件一定的情况下，𝜶 和 𝜷 不可能同时减 小，只能在发生两类错误中找一个平衡。而统计学
家认为发生第一类错误的严重程度更高，所以假设 检验中一般先控制第一类错误𝜶
### 样本容量
> 样本容量和两类错误的关系是在其他条件不变的情况下，样本容量越大，样本均值分布就越集 中，从图上反映就是正态分布的钟形曲线越瘦。 当样本容量越大， 𝜶 和 𝜷 会同时减小。换句话 说，在其他条件一致的情况下，样本量越大，实验的结果就越可信
> 
### A/B测试常常陷⼊入的误区
> 1. 统计显著性指 A 组和 B 组实验指标差异很大。×
Size of Effect 表示两组指标差异很大，而统计显著性指得到这个差异的可信程度。√
> 2. 当看到 p-value < 0.05 就马上停止实验得出结论。×
实验起初对照组相对于实验组常常 lift(提升)很大，且 p-value 很小，而这个提升有可能是由于 novel effect 引起的，比如你新上线一个功能可能第 一天吸引了一部分感兴趣的用户而过几天大家都习 惯了，才转化率回归正常，因此在实验初期就草草下结论很有可能因达不到足够的样本量造成实 验 under-power.
正确的做法是在实验还没有开始之前就计算出实验所需的最小样本量继而估算出实验天数，如果没有 达到预估实验时间不能停止实验。
> 3. 没有预估样本量就开始实验。×  
错误愿意和第 2 点类似，需要在实验还没有开始之前就预估最小样本量。
> 4. 实验没做满一个周期。×  
因为我们关注的数据，e.g., 订单量，常常会受到很多混合因素影响，这些混合因素包括天气、时间周期等等。所以为了控制这些混合因素发生的作 用，实验最好 做满整周的时间。√
> 5. A、B 两组的样本量不一致。 - 不是最好的
当 A 组和 B 组样本量不一致的时候所需要的实验时间相对于样本量一致的时候更长。为了使实验时间 在保证最小样本量的前提下尽可能更快，A 组和 B 组的样本量最好一致。
> 6. 接下来讲最重要的一个误区是很多人常常在 A/B 测试之前没有明确的实验方案。
在实验还没有开始前就需要一个明确的实验问题，实验假设，和需要实验能帮助你检测到 A 组和 B 组 有多大的不同。A/B 测试是策略和策略之间 的对比，并且 A 组和 B 组唯一不同的变量是你最关注的变 量。如果你没有明确的实验策略，那么 A/B 是不能主动帮你发现的。

## 卡放检验
### 卡方检验应用场景
`最常见的用途,考察某无序分类变量各水平在两组或多组间的分布是否一致（若事有序分类变量，使用卡方并不合适）`

- 卡方检验可以检验男性或者女性对线上买生鲜食品有没有区别；
- 不同城市级别的消费者对买SUV车有没有什么区别；
- 不同收入级别家庭的轿车拥有率比较
- 连续变量与分类变量的相关性就可以用卡方检验
### 卡方检验的基本思想
首先假设观察频数与期望频数没有差别（即：H0是观察频数与期望频数无差别，H1是观察频数与期望频数有差别）；统计量卡方值表示观察值与磷脂之间的偏离程度，
### 若想深究哪一个分类变量对连续比变量的影响大，应该看什么指标
- RR(Relative Risk)和OR(Odds Ratio)
- RR（相对危险度）：是指实验组人群中反应阳性概率与对照组人群反应阳性概率的比值，反应的是是哑巴因素与反应阳性的关联程度

- OR（优势比）：是反应阳性人群中实验因素有无的比例与反应阴性人群中实验因素有无的比例之比，

## 相关分析
### 连续变量的相关指标
> pearson相关系数表示相关性的大小（只适用于两变量呈线性相关时，若呈现曲线相关，则大小并不能代表其相关性的强弱）
> pearson相关系数计算公式主要考虑的是两个点：（1）同一个索引下，两变量数据变化的方向是否相同（2）同一个索引下，两变量距离各自平局值的大小是否同增长/减少，若满足条件是&是，则数据达到正值最大
> 相关系数的检验是基于t检验
### 有序变量的相关指标
> 检测的是一致性，一致性高就是指行变量等级高的列变量等级也高,行变量等级低的列变量等级也低，
### 非有序变量（即名义变量）的相关指标
> 卡方检验
### pearson相关系数相关记忆点
> 只适用于两变量呈线性相关时，若呈现曲线相关，则大小并不能代表其相关性的强弱
> 样本中存在的极端值对指标的影响极大，必要时可以剔除，或者可以进行变量变换，避免造成错误结论
> 可以画散点图观察数据分布，判断是否有相关趋势/是否有极值/是否线性相关等
> 要求点：1）：线性相关；2）：极值影响比较大，要做处理；3）：相应的变量呈双变量正态分布（1和2条件必要，3条件比较宽松）
### Spearman等级相关系数
> 即秩相关系数,优点对原始变量的分布不做要求，原始变量是等级资料，也可以计算秩相关系数；对服从pearson相关系数的数据（可以理解是正态分布）也可以计算秩相关系数，但统计效能比pearson相关系数要低（不容易检测出两者事实上存在的相关关系）
> 什么时候使用Spearman相关分析：当变量不满足线性相关分析的适用条件（一般是不满足正态分布），则可以使用Spearman秩相关分析
> 适用于有序分类变量
### kendall等级相关系数
> 用于反映分类变量相关性的指标，适用于两个变量均为有序分类的情况
### 偏相关分析
> 指在相关的基础上考虑两个因素和意外的各种影响因素，或者说在扣除了其它因素的作用大小以后，重新来考察这两个因素间的关联程度，

## 0-正态性检验
**正态性检验**是检验数据是否符合正态分布，也是很多统计建模的必要步骤，在Python中实现正态性检验可以使用W检验(SHAPIRO-WILK TEST)

**H0假设**：样本服从正态分布

```python
from scipy.stats import shapiro
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd
data = [0.86, 0.78, 0.83, 0.84, 0.77, 0.84, 0.81, 0.84, 0.81, 0.81, 0.80, 0.81,
       0.79, 0.74, 0.82, 0.78, 0.82, 0.78, 0.81, 0.80, 0.81, 0.74, 0.87, 0.78]
stat, p = shapiro(data)
print("stat为：%f" %stat,"p值为：%f" %p)
print("结果解释：当p值小于某个显著性水平α(如0.05)时，则认为样本不是来自正态分布的总体，否则承认样本来自正态分布的总体")
print('分布图如下')
fig, ax=plt.subplots(figsize=(5,5))
ax.hist(data,bins=5,density=True,facecolor='g',alpha=0.8)
plt.show()
```

## 相关性检验：皮尔逊相关系数

**皮尔逊相关系数（Pearson Correlation Coefficient）** 用于衡量两个变量之间的线性相关相关关系

**H0假设**：两个变量不相关

```python
#########示例代码
from scipy.stats import pearsonr
data1 = [23,20,18,29,43,35,32,40,29,26,24,26]
data2 = [1000,1000,500,500,500,100,100,100,100,100,100,100]
corr,p = pearsonr(data1,data2)
print("corr为：%f" %corr,"p值为：%f" %p)
print('散点图如下')
fig, ax=plt.subplots(figsize=(5,5))
ax.scatter(data1,data2)
plt.show()
```

## 卡方检验

**卡方检验**是一种用途很广的计数资料的假设检验方法。它属于非参数检验的范畴，卡方检验就是统计样本的实际观测值与理论推断值之间的偏离程度，实际观测值与理论推断值之间的偏离程度就决定卡方值的大小，卡方值越大，二者偏差程度越大；反之，偏差越小；若两个值完全相等时，卡方值就为0，表明理论值完全符合。

**H0假设**：观察频数与期望频数无显著差异


```python
########示例代码
from scipy.stats import chi2_contingency
import numpy as np
kf_data = np.array([[20,21], [22,24]])
kf = chi2_contingency(kf_data)
print('chisq-statistic=%.4f, p-value=%.4f, df=%i expected_frep=%s'%kf)
```

## T检验：单样本T检验

**单样本t检验**是样本均值与总体均值的比较问题。其中总体服从正态分布，从正态总体中抽样得到n个个体组成抽样样本，计算抽样样本均值和标准差，判断总体均值与抽样样本均值是否相同。

**H0假设**：样本均值无差异(μ=μ0)

**Python命令**：.ttest_1samp(data,1)

**结果解释**：当p值小于某个显著性水平α(比如0.05)时，则认为样本均值存在显著差异，具体的分析要看所选择的是双边假设还是单边假设（又分小于和大于）注意ttest_1samp进行双侧检验。

```python
from scipy import stats
data = [23,20,18,29,43,35,32,40,29,26,24,26]
print(stats.ttest_1samp(data,1)) ## 默认是双边，除2则为单边值

from statsmodels.stats import weightstats
print(weightstats.stats.ttest_1samp(data,1))
```

## T检验：两样本T检验

**两样本t检验**是比较两个样本所代表的两个总体均值是否存在显著差异。除了要求样本来自正态分布，还要求两个样本的总体方差相等也就是“方差齐性”。

**检验原假设**：样本均值无差异(μ=μ0)

**Python命令**：stats.ttest_ind(data1,data2)

当不确定两总体方差是否相等时，应先利用levene检验检验两总体是否具有方差齐性stats.levene(data1,data2)如果返回结果的p值远大于0.05，那么我们认为两总体具有方差齐性。如果两总体不具有方差齐性，需要加上参数equal_val并设定为False，如下。
**结果解释**：当p值小于某个显著性水平α(比如0.05)时，则认为样本均值存在显著差异，具体的分析要看所选择的是双边假设还是单边假设（又分小于和大于）注意stats.ttest_ind进行双侧检验。

```python
from scipy import stats
data1 = [23,20,18,29,43,35,32,40,29,26,24,26]
data2 = [1000,1000,500,500,500,100,100,100,100,100,100,100]
print(stats.ttest_ind(data1,data2))
print(weightstats.ttest_ind(x1=data1,x2=data2,alternative='two-sided'))
```

## T检验：配对T检验

**配对样本均数T检验**简称配对T检验(paired t test), 又称非独立两样本均数t检验，适用于配对设计计量资料均数的比较，其比较目的是检验两相关样本均数所代表的未知总体均数是否有差别。与独立样本T检验相比，配对样本T检验要求样本是配对的。两个样本的样本量要相同；样本先后的顺序是一一对应的。

**基本假定：**
每个样本中的观察是独立同分布的
每个样本的观察都是正态分布的
每个样本中的观察具有相同的方差
每个样本的观察结果是成对的
**检验原假设**：样本均值无差异(μ=μ0)

**Python命令**：stats.ttest_rel(data1,data2)

**结果解释**：当p值小于某个显著性水平α(比如0.05)时，则认为样本均值存在显著差异，具体的分析要看所选择的是双边假设还是单边假设（又分小于和大于）注意stats.ttest_rel进行双侧检验

```python
####示例代码，配对样本T检验
from scipy import stats
data1 = [23,20,18,29,43,35,32,40,29,26,24,26]
data2 = [1000,1000,500,500,500,100,100,100,100,100,100,100]
stats.ttest_rel(data1,data2)
```

## 方差分析（ANOVA）-F检验
### 简单方差分析
> 检验两个以上的群体在一个因素或一个维度上的均值差异
F统计量是检验“群体之间有差异”假设的检验统计量，F=MS_between/MS_within
MS_between:组间差异由于分组导致的差异
MS_within:组内差异由于随机因素导致的差异

F检验：H0是各组间均值相等

### 析因方差分析
> 处理多个群组在一个以上因素的
