## 0-Ê≠£ÊÄÅÊÄßÊ£ÄÈ™å
**Ê≠£ÊÄÅÊÄßÊ£ÄÈ™å**ÊòØÊ£ÄÈ™åÊï∞ÊçÆÊòØÂê¶Á¨¶ÂêàÊ≠£ÊÄÅÂàÜÂ∏ÉÔºå‰πüÊòØÂæàÂ§öÁªüËÆ°Âª∫Ê®°ÁöÑÂøÖË¶ÅÊ≠•È™§ÔºåÂú®Python‰∏≠ÂÆûÁé∞Ê≠£ÊÄÅÊÄßÊ£ÄÈ™åÂèØ‰ª•‰ΩøÁî®WÊ£ÄÈ™å(SHAPIRO-WILK TEST)

**H0ÂÅáËÆæ**ÔºöÊ†∑Êú¨Êúç‰ªéÊ≠£ÊÄÅÂàÜÂ∏É

```python
from scipy.stats import shapiro
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd
data = [0.86, 0.78, 0.83, 0.84, 0.77, 0.84, 0.81, 0.84, 0.81, 0.81, 0.80, 0.81,
       0.79, 0.74, 0.82, 0.78, 0.82, 0.78, 0.81, 0.80, 0.81, 0.74, 0.87, 0.78]
stat, p = shapiro(data)
print("stat‰∏∫Ôºö%f" %stat,"pÂÄº‰∏∫Ôºö%f" %p)
print("ÁªìÊûúËß£ÈáäÔºöÂΩìpÂÄºÂ∞è‰∫éÊüê‰∏™ÊòæËëóÊÄßÊ∞¥Âπ≥Œ±(Â¶Ç0.05)Êó∂ÔºåÂàôËÆ§‰∏∫Ê†∑Êú¨‰∏çÊòØÊù•Ëá™Ê≠£ÊÄÅÂàÜÂ∏ÉÁöÑÊÄª‰ΩìÔºåÂê¶ÂàôÊâøËÆ§Ê†∑Êú¨Êù•Ëá™Ê≠£ÊÄÅÂàÜÂ∏ÉÁöÑÊÄª‰Ωì")
print('ÂàÜÂ∏ÉÂõæÂ¶Ç‰∏ã')
fig, ax=plt.subplots(figsize=(5,5))
ax.hist(data,bins=5,density=True,facecolor='g',alpha=0.8)
plt.show()
```

## Áõ∏ÂÖ≥ÊÄßÊ£ÄÈ™åÔºöÁöÆÂ∞îÈÄäÁõ∏ÂÖ≥Á≥ªÊï∞

**ÁöÆÂ∞îÈÄäÁõ∏ÂÖ≥Á≥ªÊï∞ÔºàPearson Correlation CoefficientÔºâ** Áî®‰∫éË°°Èáè‰∏§‰∏™ÂèòÈáè‰πãÈó¥ÁöÑÁ∫øÊÄßÁõ∏ÂÖ≥Áõ∏ÂÖ≥ÂÖ≥Á≥ª

**H0ÂÅáËÆæ**Ôºö‰∏§‰∏™ÂèòÈáè‰∏çÁõ∏ÂÖ≥

```python
#########Á§∫‰æã‰ª£Á†Å
from scipy.stats import pearsonr
data1 = [23,20,18,29,43,35,32,40,29,26,24,26]
data2 = [1000,1000,500,500,500,100,100,100,100,100,100,100]
corr,p = pearsonr(data1,data2)
print("corr‰∏∫Ôºö%f" %corr,"pÂÄº‰∏∫Ôºö%f" %p)
print('Êï£ÁÇπÂõæÂ¶Ç‰∏ã')
fig, ax=plt.subplots(figsize=(5,5))
ax.scatter(data1,data2)
plt.show()
```

## Áõ∏ÂÖ≥ÊÄßÊ£ÄÈ™åÔºöÊñØÁöÆÂ∞îÊõºÁõ∏ÂÖ≥Á≥ªÊï∞

**ÊñØÁöÆÂ∞îÊõºÁõ∏ÂÖ≥Á≥ªÊï∞(SPEARMAN‚ÄôS RANK CORRELATION)** ÂèàÁß∞‰∏∫ÊñØÁöÆÂ∞îÊõºÁ≠âÁ∫ßÁõ∏ÂÖ≥Á≥ªÊï∞„ÄÇÊòØ‰∏ÄÁßçÈùûÂèÇÊï∞ÊñπÊ≥ïÔºåË°°Èáè‰∏§‰∏™ÂèòÈáèÁöÑ‰æùËµñÊÄßÁöÑÈùûÂèÇÊï∞ÊåáÊ†á„ÄÇ

**H0ÂÅáËÆæ**Ôºö‰∏§‰∏™ÂèòÈáè‰∏çÁõ∏ÂÖ≥


```python
#######Á§∫‰æã‰ª£Á†Å
from scipy.stats import spearmanr
data1 = [23,20,18,29,43,35,32,40,29,26,24,26]
data2 = [1000,1000,500,500,500,100,100,100,100,100,100,100]
corr, p= spearmanr(data1, data2)
print("corr‰∏∫Ôºö%f" %corr,"pÂÄº‰∏∫Ôºö%f" %p)
```
